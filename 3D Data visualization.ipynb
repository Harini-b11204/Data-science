```python
import geopandas as gp
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
npoints=100
```

```python
geo=gp.points_from_xy(np.random.uniform(low=-180,high=180,size=npoints),np.random.uniform(low=-90,high=90,size=npoints))
```

```python
df=gp.GeoDataFrame(geometry=geo)
```

```python
df['size_cloumns']=np.random.uniform(low=5,high=50,size=npoints)
```

```python
df.plot(marker='o',color='red',markersize=df['size_cloumns'],alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('GDP Map Visualization')
plt.show()
```
