```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv(r"C:\Users\harin\Downloads\Mall_Customers - Mall_Customers (1).csv")
```

```python
x=df.iloc[:,3:5].values
```

```python
s=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    k.fit(x)
    s.append(k.inertia_)
```

```python
plt.plot(range(1,11),s)
plt.title("The Elbow graph")
plt.xlabel("No of cluster")
plt.ylabel("Cluster")
plt.show()
```

```python
k=KMeans(n_clusters=5,init="k-means++",max_iter=300,n_init=10,random_state=0)
```

```python
y=k.fit_predict(x)
```

```python
color=['red','blue','green','cyan','magenta']
label=['Standard','Careless','Target','Sensible','Careful']
```

```python
for i in range(5):
    plt.scatter(x[y==i,0],x[y==i,1],c=color[i],label=label[i])
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],c='yellow',label='Centroids')
plt.title("Cluster by customers")
plt.xlabel("Annual income")
plt.ylabel("Spending money")
plt.legend()
plt.show()
```
