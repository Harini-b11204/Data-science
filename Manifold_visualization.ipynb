```python
from yellowbrick.features import Manifold
from yellowbrick.datasets import load_concrete
```

```python
x,y=load_concrete()
```

**Create Map using Manifold**

```python
v=Manifold(manifold="isomap",n_neighbors=15)
v.fit_transform(x,y)
v.show()
```
